```python
#All Models in this program actualSales the validity of the model at predicting actiual values.  
# I have not yet added actualSales/forecasting functionality.  I will do one week's work of actualSales at a time.  


import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import statsmodels.api as sm
import operator
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
#A vectorized mape function for efficiency

def mape_vectorized_v2(a, b): 
    mask = a != 0
    return (np.fabs(a - b)/a)[mask].mean() 
```

```python
#Creating the Dataframe

df = pd.read_csv('/Users/jnapolitano/projects/CoxOil/potLog6.csv')
```

```python
#creating the dictionary to hold the Errors of each method.  Will find minimum(best) value at the end
rmseDictionary = {}

mapeDictionary = {}
```

```python
#Rearanging data into two new data frames aggregated by the sums of days
#historic data spans 1.1.2020 to 3.22.2020
#actualSales data is the week of march 23

allData = df[:984].copy()
allData['Timestamp'] = pd.to_datetime(allData.date,format='%Y-%m-%d')
allData.index = allData.Timestamp
allData = allData.resample('D').sum()
allData = allData.drop(columns=['hour', 'week'])

historic = df[:984].copy()

historic['Timestamp'] = pd.to_datetime(historic.date,format='%Y-%m-%d')
historic.index = historic.Timestamp
historic = historic.resample('D').sum()
historic = historic.drop(columns=['hour', 'week'])

actualSales = df[984:1067].copy()

actualSales['Timestamp'] = pd.to_datetime(actualSales.date,format='%Y-%m-%d')
actualSales.index = actualSales.Timestamp
actualSales = actualSales.resample('D').sum()0
actualSales = actualSales.drop(columns=['hour', 'week'])
```

```python
#Plotting the Historic and actualSales data on the same plane

historic.sales.plot(figsize=(15,8), title= 'Sales', fontsize=14)
actualSales.sales.plot(figsize=(15,8), title= 'Sales', fontsize=14)

plt.show()
```

```python
#BUilding the predicted_results able which will hold all predicted values and be plotted

dd= np.asarray(actualSales.sales)
predicted_results = actualSales.copy()
```

```python
#the Sarima Model is another seasonal model. 
#Fitting

fit1 = sm.tsa.statespace.SARIMAX(historic.sales, order=(2, 1, 4),seasonal_order=(0,1,1,7)).fit()
predicted_results['SARIMA'] = fit1.forecast(len(actualSales.sales), dynamic=True)
plt.figure(figsize=(16,8))
plt.plot( historic['sales'], label='Historic')
plt.plot(actualSales['sales'], label='Actual Sales')
plt.plot(predicted_results['SARIMA'], label='SARIMA')
plt.legend(loc='best')
plt.show()

SarimaParameters = fit1.params
rmse = sqrt(mean_squared_error(actualSales.sales, predicted_results.SARIMA))
rmseDictionary["SARIMA"] = rmse

mape = mape_vectorized_v2(actualSales.sales, predicted_results['SARIMA'])
mapeDictionary['SARIMA'] = mape 

#print("The RMSE for the SARIMA Model is equal to {}\n".format(rmse))
```

```python
#HoltWinter Model 
model = ExponentialSmoothing(np.asarray(historic['sales']) ,seasonal_periods=7 ,trend='add', seasonal='add',)
fit1 = model.fit()
predicted_results['Holt_Winter'] = fit1.forecast(len(actualSales))

HoltWinterParameters = fit1.params


rmse = sqrt(mean_squared_error(actualSales.sales, predicted_results.Holt_Winter))
rmseDictionary["Holt_Winter"] = rmse

mape = mape_vectorized_v2(actualSales.sales, predicted_results['Holt_Winter'])
mapeDictionary['Holt_Winter'] = mape 

plt.figure(figsize=(16,8))
plt.plot( historic['sales'], label='Historic')
plt.plot(actualSales['sales'], label='Actual Sales')
plt.plot(predicted_results['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()


```
